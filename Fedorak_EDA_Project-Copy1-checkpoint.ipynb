We can notice very dramatic difference in the scale of the "Rate" axis between our male and female data.
```python
male_only2000 = male_only[male_only['Year'] == '2000']
male_only2000
```

```python
countries = male_only2000['Country']
male_rates = male_only2000['Rate']
female_rates = female_only2000['Rate']

bar_width = 0.35
index = range(len(countries))

fig, ax = plt.subplots()
bar1 = ax.bar(index, male_rates, bar_width, label='Male Rate')
bar2 = ax.bar([i + bar_width for i in index], female_rates, bar_width, label='Female Rate')

ax.set_xlabel('Country')
ax.set_ylabel('Rate')
ax.set_title('Murder Rates for Males and Females (2000-2020)')
ax.set_xticks([i + bar_width / 2 for i in index])
ax.set_xticklabels(male_only2000['Country'])
ax.legend()
plt.xticks(rotation=55)

plt.show()
```

```python
import pandas as pd
from matplotlib import pyplot as plt

year = colombia_onlym['Year'].head(21)
rate = colombia_onlym['Rate'].head(21)

fig, ax = plt.subplots(figsize =(16, 9))

ax.barh(year, rate)

for s in ['top', 'bottom', 'left', 'right']:
    ax.spines[s].set_visible(False)

ax.xaxis.set_ticks_position('none')
ax.yaxis.set_ticks_position('none')

ax.xaxis.set_tick_params(pad = 5)
ax.yaxis.set_tick_params(pad = 10)
 
ax.grid(color ='grey',
        linestyle ='-.', linewidth = 0.5,
        alpha = 0.2)
 
ax.invert_yaxis()

for i in ax.patches:
    plt.text(i.get_width()+0.2, i.get_y()+0.5, 
             str(round((i.get_width()), 2)),
             fontsize = 10, fontweight ='bold',
             color ='grey')

ax.set_title('Rate of Intentional Homicide in Males in Colombia from 2000-2020', loc ='left', )
plt.xlabel("Rate")
plt.ylabel("Year")

plt.show()
```

Finally, let's answer our question about whether intentional homicide rates of females differs between countries with strong women's rights and countries lacking in women's rights. For countries who have strong women's rights, we will use Norway, Switzerland, and the Netherlands. For countries lacking in women's rights, we will use India, Colombia, and Brazil. Although these three countries are not known for their lack of women's rights, it should be noted that they are the countries within our data which are the lowest on the list of the 2018 Women, Peace and Security Index.  https://www.peacewomen.org/resource/and-worst-country-be-woman

```python
top3countries = ['Norway', 'Switzerland', 'Netherlands']

datatop3 = female_only[female_only['Country'].isin(top3countries)]

plt.figure(figsize=(5, 3))
sns.barplot(x='Country', y='Rate', data=datatop3, palette='Set1')

plt.xlabel('Country')
plt.ylabel('Rate')
plt.title("Rates of Intentional Homicide in Females in Countries with Strong Women's Rights")

plt.show()
```

```python
bottom3countries = ['India', 'Colombia', 'Brazil']

databottom3 = female_only[female_only['Country'].isin(bottom3countries)]

plt.figure(figsize=(5, 3))
sns.barplot(x='Country', y='Rate', data=databottom3, palette='Set1')

plt.xlabel('Country')
plt.ylabel('Rate')
plt.title("Rates of Intentional Homicide in Females in Countries with Weak Women's Rights")

plt.show()
```

If we compare our top 3 women's rights countries to our bottom 3 women's rights countries, there is an obvious difference when looking at the scales of these two graphs. The rate of intentional homicide in females in the countries with poor women's rights is significantly higher, ranging up to 6.00. Conversely, in our data showing the top 3 countries with strong women's rights, we see that the scale only ranges up to about 1.10. This dramatic difference may be due to how women are treated and looked at by society in these countries.
